### Good way to write classes

```python
class Bus:
  #constructor->
  def __init__(self,windows,tyres,engine):   #windows,tyres,engine will pass as parameters
       self.windows=windows     #here in self.windows this windows will be name of attributes
       self.tyres=tyres
       self.engine=engine

  #Self is used because with using(through) self we can directly call any attributes that are inside
  #car class

  def self_driving(self):      #method
    print("the car type is {} engine".format(self.engine))

  def self_driving1(self,engine):
    print("the car type is {} engine".format(engine))

bus1= Bus(4,4,"petrol")
print("The no of tyres in object bus1 is {}".format(bus1.tyres))
print("The no of windows in object bus1 is {}".format(bus1.windows))
bus1.self_driving()
bus1.self_driving1("electric")
```

###Multiple constructor in Python
-Not allowed in Python(but allow in java)

-Last constructor will override other


```python
class Animal:
  def __init__(self,name,species):
    self.name=name
    self.species=species

  def __init__(self,name,species,age):
    self.name=name
    self.species=species
    self.age=age

  def make_sound(self,sound):
    return "The animal is {} and says {}".format(self.name,sound)

dog=Animal("dog","mammals")
print(dog)

#dog1=Animal("dog","mammals",10)
#print(dog1)
```

-But there is an Hack we write multiple constructor in python using positional arguments  

```python
class Ani:
  def __init__(self,*args):
    if len(args)==1:
      self.name=args[0]
    elif len(args)==2:
      self.name=args[0]
      self.species=args[1]
    elif len(args)==3:
      self.name=args[0]
      self.species=args[1]
      self.age=args[2]

  def make_sound(self,sound):
    return "The animal is {} and says {}".format(self.name,sound)

dog=Ani("dog","mammals")
print(dog.name)
print(dog.species)
```

###Inheritance in python

```python
class Car2:
  def __init__(self,windows,doors,enginetype):
    self.windows=windows
    self.doors=doors
    self.enginetype=enginetype

  def driving(self):
    print("Car is used for driving")

class Audi(Car2):
  def __init__(self,windows,doors,enginetype,horsepower):
    super().__init__(windows,doors,enginetype)
    self.horsepower=horsepower

  def selfdriving(self):
    print("It is a self driving car")

audiq7=Audi(4,5,"Diesel",200)

print(audiq7.enginetype)
audiq7.driving()
audiq7.selfdriving()

print(dir(audiq7))
```

###Python Data Classes


In Python, the dataclass decorator is a feature introduced in Python 3.7 that provides a concise way to define classes primarily intended to store data. It automatically generates several special methods, such as init, repr, and eq, based on the class attributes you define. This simplifies the process of creating and working with data-focused classes.


```python
from dataclasses import dataclass
@dataclass
class Person:
    name:str
    age:int
    profession:str
```

The @dataclass decorator automatically generates the following methods for you:

--init(): Initializes the object and assigns the provided values to the attributes.

--repr(): Provides a string representation of the object.

--eq(): Implements equality comparison between two objects of the class based on their attributes.


```python
person1=Person('ABC',17,'SE')
print(person1)
person1.age
```

```python
@dataclass
class Rectangle:
    width:int
    height:int
    color:str='white'
```

```python
rectangle1=Rectangle(12,14)
rectangle2=Rectangle(13,14,'red')
rectangle2.color
```

###Encapsulation In Python


Encapsulation is a fundamental principle in object-oriented programming that focuses on bundling data and the methods that operate on that data into a single unit called a class. It allows you to control the access and visibility of the data and methods, providing a way to protect and organize your code



```python
### Access Modifier--->Encapsulation
## private

class Person:
    ## constructor
    def __init__(self,name,age):
        self.__name=name     #__name is a private access modifier(can't be access directly outside class)
        self.__age=age

    def display_info(self):
        print(f"the person name is {self.__name} and the age is {self.__age}")
```

```python
person=Person("Krish",32)
person.__name
```

```python
### Access Modifier--->Encapsulation
## Protected :It can only be used in derived class.

class Person:
    ## constructor
    def __init__(self,name,age):
        self._name=name
        self._age=age
```

```python
person=Person("Krish",32)
```

```python
dir(person)
person._age  #actually we use it outsid class but again wrong technique because python never say it's compulsory .It a way if you want to folllow follow it or not.
```

```python

class Student(Person):
    def __init__(self,name,age):
        super().__init__(name,age)

    def display_info(self):
        print(f"the person name is {self._name} and the age is {self._age}")
```

```python
student1=Student("Krish",33)
student1.display_info()
```
